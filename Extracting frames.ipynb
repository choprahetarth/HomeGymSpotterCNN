```python
import cv2
vidcap = cv2.VideoCapture('WhatsApp Video 2020-05-08 at 11.50.39 AM.mp4')
import os

os.chdir('frames')
```

```python

success,image = vidcap.read()
count = 0

while success:
  cv2.imwrite("4frame%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```

```python
import glob
fname = (glob.glob("4*.jpg"))

for f in fname:
    i = cv2.imread(f)
    i = cv2.rotate(i, cv2.ROTATE_90_CLOCKWISE)
    cv2.imwrite(f, i)
```
